```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#get ranked data of 33 volunteers 
lc = pd.read_csv("./loc.csv",sep=',')
bf = pd.read_csv("./rankedbff.csv",sep=',')
```

```python
lc
```

```python
type(lc)
```

```python
bf
```

```python
lc =lc.T
bf =bf.T
```

```python
lc
```

```python
bf
```

```python
lc_s = lc.corr(method='spearman')
bf_s = bf.corr(method='spearman')
```

```python
lc_s
```

```python
bf_s
```

```python
x = []
y = []
for i in range(0,33):
    for j in range(i+1,33):
        y.append([bf_s[i][j]])
        x.append([lc_s[i][j]])
```

## Data visualization

```python
plt.scatter(x,y)
plt.title("Location and Personality")
plt.xlabel("location coefficient")
plt.ylabel("personality coefficient")
plt.axis([-1.0,1.0,-1.0,1.0])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X=pd.DataFrame(x),y=pd.DataFrame(y))
```

```python
plt.scatter(x,y)
plt.title("Location and Personality")
plt.xlabel("location coefficient")
plt.ylabel("personality coefficient")
plt.axis([-1.0,1.0,-1.0,1.0])
plt.grid(True)
plt.plot(x,model.predict(pd.DataFrame(x)),color='red')
plt.show()
```

## Results

```python
#선형 회귀 선의 그래디언트 
print(model.coef_[0][0],2)
```

```python
#그래프의 x값이 0인 경우 y값 예측하여 y절편(0.06)을 얻음 
print(model.intercept_[0],2)
```

```python
#훈련된 모델 기반으로 location coefficient이 0.5일때 personality coefficient는 0.14로 예측됨 
print(model.predict([[0.5]])[0][0],2)
```
